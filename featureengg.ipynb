```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
def tenure_group(tenure):
    if tenure <= 12:
        return "New"
    elif tenure <= 36:
        return "Regular"
    elif tenure <= 60:
        return "Loyal"
    else:
        return "Champion"

df['tenure_group'] = df['tenure'].apply(tenure_group)
```

```python
df['tenure_group']
```

```python
def monthly_charge_segment(MonthlyCharges) :
  if(MonthlyCharges)<30 :
    return "Low"
  elif(MonthlyCharges>30 and MonthlyCharges<70) :
    return "Medium"
  elif(MonthlyCharges>70):
    return "High"
  else :
    return "Invalid"





df['monthly_charge_segment']=df['MonthlyCharges'].apply(monthly_charge_segment)
```

```python
df['monthly_charge_segment']
```

```python
def has_Internet_Service(InternetService) :
  if(InternetService=="DSL" or InternetService=="Fiber optic") :
    return 1
  else :
    return 0






df['has_Internet_Service']=df['InternetService'].apply(has_Internet_Service)
```

```python
def is_multi_line_user(MultipleLines):
  if MultipleLines=="Yes":
    return 1
  else :
    return 0





df['is_multi_line_user']=df['MultipleLines'].apply(is_multi_line_user)
```
